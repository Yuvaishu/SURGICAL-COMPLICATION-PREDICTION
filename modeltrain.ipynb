```python
from tensorflow.keras.layers import Input, Dense, Activation, Add
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
def res_block(x, filters, block_num):
    # Define a residual block with skip connections
    y = Dense(filters, activation='relu', name=f'res{block_num}_a')(x)
    y = Dense(filters, activation='relu', name=f'res{block_num}_c')(y)
    y = Dense(filters, activation='relu', name=f'res{block_num}_b')(y)
    y = Add(name=f'res{block_num}_add')([x, y])
    y = Activation('relu', name=f'res{block_num}_out')(y)
    return y

inputs = Input(shape=(X.shape[1],))
x = Dense(164, activation='relu')(inputs)
x = res_block(x, 164, 1)
x = res_block(x, 164, 2)
x = res_block(x, 164, 3)
x = Dense(len(set(y)), activation='softmax')(x)

model = Model(inputs=inputs, outputs=x)

# Compile the model
model.compile(optimizer=Adam(lr=0.01), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=5000, batch_size=32)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Model accuracy:", accuracy)

```
